```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
PJDF = pd.read_pickle("./DeliveryDataFrame.pkl")
```

```python
plt.figure(0)
plt.title("Papa Johns Earnings and Miles Driven in Summer 2020")
plt.plot(PJDF.index, PJDF.Totals, color="blue")
plt.xticks(PJDF.index[::3], rotation=300)
plt.ylabel('Dollars ($)', color="blue")
plt.xlabel('Work Day (MM/DD/YY)')
plt.twinx()
plt.plot(PJDF.index, PJDF.MilesDriven, color="purple")
plt.xticks(PJDF.index[::3], rotation=300)
plt.ylabel('Miles Driven (miles)', color="purple")
#####
plt.figure(1)
plt.title("Papa Johns Estimated Hourly Rate and Miles Driven in Summer 2020")
plt.plot(PJDF.index, PJDF.PayRate, color="g")
plt.xticks(PJDF.index[::3], rotation=300)
plt.ylabel('Pay Rate ($ / hour)', color="g")
plt.xlabel('Work Day (MM/DD/YY)')
plt.hlines(PJDF['PayRate'].mean(), PJDF.index[0], PJDF.index[-1], colors='black', linestyles='dashdot', label='Average Pay Rate')
plt.legend(loc="upper left")
plt.twinx()
plt.plot(PJDF.index, PJDF.MilesDriven, color="purple")
plt.xticks(PJDF.index[::3], rotation=300)
plt.ylabel('Miles Driven (miles)', color="purple")
######
plt.figure(2)
plt.title("Papa Johns Total Hours worked and Miles Driven in Summer 2020")
plt.plot(PJDF.index, PJDF.TotalHours, color="red")
plt.xticks(PJDF.index[::3], rotation=300)
plt.ylabel('Total Hours (hours)', color="red")
plt.xlabel('Work Day (MM/DD/YY)')
plt.twinx()
plt.plot(PJDF.index, PJDF.MilesDriven, color="purple")
plt.xticks(PJDF.index[::3], rotation=300)
plt.ylabel('Miles Driven (miles)', color="purple")

```
